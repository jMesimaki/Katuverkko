```python
import pandas as pd
import numpy as np
```

```python
onn = pd.read_csv("C:/data/L3Pilot/ScriptTest/Pyth/OnntestCSV.csv", sep=";")
```

```python
VRUs = onn["lkmjk"] + onn["lkmpp"]
onn["VRU"] = VRUs
```

```python
grouped = onn.groupby("ID")
```

```python
grouped.get_group(1)
```

```python
idvalues = onn['ID'].nunique()
idvector = list(range(1, idvalues+1))
len(idvector)
```

```python
colnames = ["ID", "Section", "Type", "Length (km)", "ADT", "All injury acc", "Head-on", "VRU", "Other"]
df = pd.DataFrame(columns = colnames)
df["ID"] = idvector  
```

```python
# Create empty vectors. These will be filled with accidents of specified type and added to final dataframe
OtherVectorFinal = []
VRUVectorFinal = []
HeadOnVectorFinal = []
AllInjuryFinal = []
ContainsVRUFinal = []
isjkpp = []
# Iterate over groups (road IDs)
for i in range(1, (idvalues+1)):
    
    # reset vectors
    OtherVector = []
    VRUVector = []
    HeadOnVector = []
    ContainsVRU = []
    
    # access road's accident types
    grp = grouped.get_group(i)
    grp = grp.reset_index()
    #grp = grp.to_frame()
    grp = grp[["onntyyppi", "lkmjk", "lkmpp", "VRU"]] 
    #print(grp)
    
    # save road's accident types to vectors [smarter way to do this?]
    occurrencesZero = np.count_nonzero(grp["onntyyppi"] == 0)
    OtherVector.append(occurrencesZero)
    occurrencesOne = np.count_nonzero(grp["onntyyppi"] == 1)
    OtherVector.append(occurrencesOne)
    occurrencesTwo = np.count_nonzero(grp["onntyyppi"] == 2)
    HeadOnVector.append(occurrencesTwo)
    occurrencesThree = np.count_nonzero(grp["onntyyppi"] == 3)
    HeadOnVector.append(occurrencesThree)
    occurrencesFour = np.count_nonzero(grp["onntyyppi"] == 4)
    OtherVector.append(occurrencesFour)
    occurrencesFive = np.count_nonzero(grp["onntyyppi"] == 5)
    OtherVector.append(occurrencesFive)
    occurrencesSix = np.count_nonzero(grp["onntyyppi"] == 6)
    OtherVector.append(occurrencesSix)
    occurrencesSeven = np.count_nonzero(grp["onntyyppi"] == 7)
    OtherVector.append(occurrencesSeven)
    occurrencesEight = np.count_nonzero(grp["onntyyppi"] == 8)
    OtherVector.append(occurrencesEight)
    occurrencesNine = np.count_nonzero(grp["onntyyppi"] == 9)
    OtherVector.append(occurrencesNine)
    
    #check for pp/jk
    for j in range(0, len(grp)):
        if (grp.at[j, "VRU"] >= 1):
            if (grp.at[j, "onntyyppi"] == 2 or grp.at[j, "onntyyppi"] == 3):
                VRUVector.append(1)
                HeadOnVector.append(-1)
            else:
                VRUVector.append(1)
                OtherVector.append(-1)
        #else:
            #isjkpp.append(0)
    totalVRUheadon = sum(isjkpp)
    VRUVector.append(totalVRUheadon)
    
    # sum accident types for road being looped
    HeadOnRoad = sum(HeadOnVector)
    VRURoad = sum(VRUVector)
    OtherRoad = sum(OtherVector)
    AllInjury = (HeadOnRoad+VRURoad+OtherRoad)
    JKPP = sum(ContainsVRU)
    
    # Add sums to final vectors
    HeadOnVectorFinal.append(HeadOnRoad)
    VRUVectorFinal.append(VRURoad)
    OtherVectorFinal.append(OtherRoad)
    AllInjuryFinal.append(AllInjury)
    ContainsVRUFinal.append(JKPP)
    

    
#Add final vectors to df
df["Head-on"] = HeadOnVectorFinal
df["VRU"] = VRUVectorFinal
df["Other"] = OtherVectorFinal
df["All injury acc"] = AllInjuryFinal

df
```
